# 6. Correlation

- Using the titanic dataframe, again, print the correlation between the 'age' and 'sex'
- En utilisant le dataframe 'titanic', encore une fois, afficher la corrélation entre 'l'âge' et le 'sexe' des passagers

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
df = pd.read_csv("titanic.csv")
df["Sex"]=encoder.fit_transform(df["Sex"])
df["Sex"].corr(df["Age"])
```

- Using the given function, print the correlation metrix of the two dataframes you have, call the instructor and comment the results you have
- En utilisant la fonction donnée, affichez la matrice de corrélation des deux dataframes que vous avez, appelez l'instructeur et commentez les résultats que vous avez

```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True)
    s = sns.heatmap(
    corr, 
    cmap = cmap,
    square=True, 
    cbar_kws={ 'shrink' : .9 }, 
    ax=ax, 
    annot = True, 
    annot_kws = { 'fontsize' : 12 }
)
```

```python
df = pd.read_csv("titanic.csv")
plot_correlation_map(df)
```

```python
df2 = pd.read_csv("auto_ds.csv")
plot_correlation_map(df2)
```
